```python
import pandas as pd
import numpy as np
```

```python
from bokeh.charts import Bar, show, output_notebook
from bokeh.layouts import row
output_notebook()
```

```python
df1 = pd.read_csv('results.csv')
del df1['Unnamed: 0']
del df1['Unnamed: 0.1']
```

```python
df1.head()
```

```python
df1['Elapsed Time'] = np.log(df1['Elapsed Time'])
```

```python
df1['Expansions'] = np.log(df1['Expansions'])
df1['Optimal Length'] = np.log(df1['Optimal Length'])
```

```python
df1.head()
```

```python
f1 = Bar(df1, label='Search ID', values='Elapsed Time', group='Problem ID', legend = True, width = 250, height = 250, ylabel = "Log(Elapsed Time)")
f2 = Bar(df1, label='Search ID', values='Expansions', group='Problem ID', legend = True, width = 250, height = 250, ylabel="Log(Expansions)")
f3 = Bar(df1, label='Search ID', values='Optimal Length', group='Problem ID', legend = True, width = 250, height = 250, ylabel="Log(Optimal Path length)")
show(row([f1, f2, f3]))
```
